## Predicting House Price (sqft_model)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
X = house_data.iloc[:,5]
X = np.array(X)
X = X.reshape(X.shape[0], 1)
y = house_data.iloc[:,2]

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=470,test_size=0.2)
sqft_model = LinearRegression()
sqft_model.fit(X_train, y_train)
predictions = sqft_model.predict(X_test)
predictions
for i,prediction in enumerate(predictions):
    print('Predicted:%3f' %(prediction))
```

## Finding out the RSS value

```python
print ('R-squared: %.4f' % sqft_model.score(X_test, y_test))
```

```python
print ('R-squared: %.4f' % sqft_model.score(X_train, y_train))
```

## Ploting the prediction

```python
import matplotlib.pyplot as plt
plt.plot(X_test,y_test,'.',X_test,predictions,'-')
```

## Predicting House Price (more_features_model)

```python
my_features=['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode',"waterfront","view","grade","sqft_above","sqft_basement","yr_built","yr_renovated","zipcode","lat","long","sqft_living15","sqft_lot15"]
```

```python
X = house_data[my_features]
y = house_data.iloc[:,2]
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0,test_size=0.2)
more_features_model = LinearRegression()
more_features_model.fit(X_train, y_train)
predictions = more_features_model.predict(X_test)
predictions
for i,prediction in enumerate(predictions):
    print('Predicted:%3f' %(prediction))
```

## Finding out the RSS value

```python
print ('R-squared: %.4f' % more_features_model.score(X_test, y_test))
```

```python
print ('R-squared: %.4f' % more_features_model.score(X_train, y_train))
```

## ploting the predictions

```python
import matplotlib.pyplot as plt
plt.plot(X_test,y_test,'.',X_test,predictions,'-')
```
